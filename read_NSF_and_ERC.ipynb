```python

dict_NSF_awardees={}
with open("list_NSF_data.txt") as fp:
    Lines = fp.readlines()
    for filename_ in Lines:
#         if re.split('=|&',filename_)[1]=='2006':
#             break        
        xml_filenames=glob.glob('NSF_extracted_XMLs/z_'+re.split('=|&',filename_)[1]+'/*')
        
        # LOOP OVER ALL AWARDEES IN THE YEAR 
        # for Z in xml_filenames:
        Z=xml_filenames[0]
        x = etree.parse(Z)
        xx= etree.tostring(x, pretty_print=True)
        xx=re.split('\n',xx.decode("utf-8") )
        dict_NSF_awardees[xml_filenames[0]]={}
        for y in xx:
            z=re.split('<|>',y)[1:-2]
            if len(z)>0:
                print(z)
                dict_NSF_awardees[xml_filenames[0]][z[0]]=z[1]
                
        print('\n------')
```
